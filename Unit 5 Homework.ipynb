```python
# Set number of Simulation
num_sims = 500

# Configure a Monte Carlo simulation to forecast 30 years for 40/60 Portfolio
MC_AGG_SPY = MCSimulation(
    portfolio_data = df_ticker,
    num_simulation = num_sims,
    num_trading_days = 252*30
)
```

```python
 # Run Monte Carlo simulations to forecast 30 years daily returns
MC_AGG_SPY.calc_cumulative_return()
```
